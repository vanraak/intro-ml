```python
#Create DataFrame (df) based on the csv file
df = pd.read_csv('data/customers.csv')
```

```python
#Display the first 5 lines of the DataFrame
df.head()
```

```python
#Create a scatter plot
x=df['Annual income (Euro x1000)']
y=df['Customer satisfaction (1-100)']

plt.scatter(x,y)
plt.xlabel("Annual income (Euro x1000)")
plt.ylabel("Customer satisfaction (1-100)")
plt.title('Scatter plot')
plt.show()
```

```python
#Import Kmeans, define and fit the model, and assign generated labels to a new column in the DataFrame (df)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5)
kmeans.fit(df)
label=kmeans.predict(df)
df['Label']=label
```

```python
#Display the first 5 lines of the DataFrame (df)
df.head()
```

```python
#Generate a new scatter plot to displays the clusters

#First create 'groups' for each cluster
cluster1=df[df['Label']==0]
cluster2=df[df['Label']==1]
cluster3=df[df['Label']==2]
cluster4=df[df['Label']==3]
cluster5=df[df['Label']==4]

#Next, plot x and y per cluster
x=cluster1['Annual income (Euro x1000)']
y=cluster1['Customer satisfaction (1-100)']
plt.scatter(x, y, color='Red')

x=cluster2['Annual income (Euro x1000)']
y=cluster2['Customer satisfaction (1-100)']
plt.scatter(x, y, color='Green')

x=cluster3['Annual income (Euro x1000)']
y=cluster3['Customer satisfaction (1-100)']
plt.scatter(x, y, color='Purple')

x=cluster4['Annual income (Euro x1000)']
y=cluster4['Customer satisfaction (1-100)']
plt.scatter(x, y, color='Blue')

x=cluster5['Annual income (Euro x1000)']
y=cluster5['Customer satisfaction (1-100)']
plt.scatter(x, y, color='Pink')

#Add a title and labels to the plot
plt.xlabel("Annual income (Euro x1000)")
plt.ylabel("Customer satisfaction (1-100)")
plt.title('Scatter plot')

#Show the plot
plt.show()
```

```python
#Elbow method:

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="blue", marker ="*")
plt.xlabel("Number of clusters")
plt.xticks([i for i in range(1,11)])
plt.ylabel("WCSS")
plt.title("Elbow method")
plt.show()
```
