```python
#Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#Create a DataFrame containing all the training observations
train = pd.read_csv('data/mnist_train.csv.gz', header=None, compression="gzip")
```

```python
#Display the first 5 rows of the DataFrame:
train.head()
```

```python
#Column 0 contains the labels. We use that to split our DataFrame into two parts: X_train and y_train:
X_train = train.drop(columns=0)
y_train = train[0]
```

```python
#We see a lot of 0s in our dataset. 0 = black, 255= white.
```

```python
#Display the first 50 numbers of the train DataFrame (X_train) and their labels (y_train):
for i in range(50):
    ax= plt.subplot(5,10 ,i+1)
    im=ax.imshow(X_train.iloc[i].values.reshape(28,28), cmap='gray')
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    plt.title(y_train.iloc[i])
```

```python
#We can display the max value of a pixel in the center of the figure:
X_train[400].max()
```

```python
#Display the first 5 rows of our X_train DataFrame:
X_train.head()
```

```python
#Display the first 5 rows of our labels (y_train)
#Note: this is a series, not a DataFrame
y_train.head()
```

```python
#Next, we load our test data into a DataFrame
test = pd.read_csv('data/mnist_test.csv.gz', header=None, compression="gzip")
```

```python
#Display the 5 first rows:
test.head()
```

```python
#The structure is the same as before. We again split the data into a train dataset and a test dataset:
X_test=test.drop(columns=0)
y_test=test[0]
```

```python
#Import the k-nearest neighbors model from Scikit-learn
from sklearn.neighbors import KNeighborsClassifier

#We create an instance of the model named 'knn', with k=3
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
#Fit the model to the data:
knn.fit(X_train, y_train)
```

```python
#Display the accuracy of the model:
knn.score(X_test, y_test)
```

```python
#Examine the True (right) and False (wrong) predictions using a Confusion Matrix:
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
```

```python
#We can also save the predicted values
y_pred=knn.predict(X_test)
```

```python
#Create a new DataFrame: df
df=pd.DataFrame()
```

```python
#Add two columns: actual labels and predicted labels:
df['actual']=y_test
df['prediction']=y_pred
```

```python
#Use np.where to create a column 'match' with a value of 1 if predicted label=true label, 0 otherwise:
df['match']=np.where(df['actual']==df['prediction'],1,0)
```

```python
#Only keep misclassified digits
df=df[df['match']==0]

#Display the first 10 cases where True label != Predicted label
df.head(10)
```

```python
#Display an incorrectly predicted number
i=195
plt.figure(figsize = (1,1))
plt.imshow(X_test.iloc[i].values.reshape(28,28), cmap='gray')
print(f'actual: {y_test.iloc[i]}')
print(f'prediction: {y_pred[i]}')
```

```python
#We can also show for example only cases where, for example, the number 7 is misclassified:
df[df['actual']==7].head(10)
```
